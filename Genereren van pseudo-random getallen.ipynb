<h2>Codeer een uniform random number generator</h2>

```python
from math import sqrt
import matplotlib.pyplot as plt
```

```python
class random_number_generator:

    def __init__(self, seed):
        self.numbers = [seed]
        self.a = 1664525
        self.c = 1013904223
        self.m = 2**32
    
    def next_int(self):
        next_int = (self.a*self.numbers[-1]+self.c)%self.m
        self.numbers.append(next_int)
        return next_int
    
    def get_numbers(self):
        return self.numbers[1:]
```

<h2>Test je random number generator</h2>

```python
RNG = random_number_generator(20)
for i in range(100):
    RNG.next_int()
```

```python
plt.plot(range(1, len(RNG.get_numbers())+1), RNG.get_numbers())
plt.title("Willekeurige getallen van de test")
plt.ylabel("gegenereerde willekeurige getal")
plt.xlabel("nde willekeurige getal")
plt.show()
```

<h2>Leg je methode en je bevindingen vast in een verslag</h2>

Voor het berekenen van een willekeurig getal heb ik gebruik gemaakt van de "The congruential method". Deze methode berekend op basis van het vorige willekeurige getal een nieuw willekeurig getal. Bij deze methode word gebruik gemaakt van de volgende formule.

xi+1 = axi + c (mod m) 

In deze formule heb je 4 parameters a, xi, c en m. Deze parameters bepalen hoe het volgend willekeurige getal berekend word. Om een willekeurig getal te kunnen berekenen is er een seed nodig, dit is xi. xi word vermeenigvuldigd met a en vervolgens word er c bij opgeteld. Tot slot word de modulo berekent van deze uitkomst. oftewel m word net zo vaak van deze uitkomst afgehaald tot dat het resultaat tussen 0 en m-1 is. Deze uitkomst is dan het volgend willekeurige getal. En dit getal kan dan weer gebruikt worden voor de berekening van het volgende willekeurige getal.

m bepaald dus de range van het willekeurige getal dat je wilt berekenen als m = 20 dan is je willekeurige getal altijd van 0 tot en met 19, oftewel van 0 tot en met m-1. Verder viel mij op dat bij deze methode je al gauw in een loop komt van dezelfde willekeurige getallen. Dan is deze generator niet echte meer willekeurig omdat je dan zou kunnen voorspellen wat het volgende getal word. <h2>Gebruik de random number generator om pi te benaderen</h2>

```python
def calculate_pi(num_dots):
    inside_circle = 0
    xy_generator = random_number_generator(50)
    for i in range(num_dots):
        x = xy_generator.next_int()
        y = xy_generator.next_int() 
        if sqrt((x/xy_generator.m)**2 + (y/xy_generator.m)**2) <= 1:
            inside_circle = inside_circle + 1
    return inside_circle/num_dots*4
```

```python
print(calculate_pi(100000))
```
