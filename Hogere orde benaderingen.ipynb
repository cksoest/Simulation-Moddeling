```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Euler</h2>

functie die de differentiaal vergelijking oplost met de methode van Euler. De aantal stappen en stap grootte zijn gegeven aan de fuctie, en de x en y as geeft de functie terug.

```python
def euler_forward_e(steps, h):
    x = np.zeros(steps+1)
    y = np.zeros(steps+1)
    x[0] = 0
    y[0] = 1
    for step in range(steps):
        x[step+1] = h * (step+1)
        y[step+1] = y[step] + h*x[step]
    return x, y
```

<h2>Heun</h2>

functie die de differentiaal vergelijking oplost met de metode van Heun. De aantal stappen en stap grootte zijn gegeven aan de fuctie, en de 
x en y as geeft de functie terug.

```python
def heun_e(steps, h):
    x = np.zeros(steps+2)
    y = np.zeros(steps+1)
    x[0] = 0
    y[0] = 1
    for step in range(steps+1):
        x[step+1] = h * (step+1)
    for step in range(steps):
        euler = y[step] + h*x[step]
        y[step+1] = y[step] + h * ((x[step] + euler) / 2)
    x = x[:-1]
    return x, y
```

<h2>Visualisatie van de oplossing van de differentiaal vergelijking, met 5 stappen van 0.2

```python
x, y = euler_forward_e(5, 0.2)
plt.plot(x,y)
plt.title("Grafiek van de formule e^x Euler")
plt.ylabel("e^x")
plt.xlabel("Gegeven x waarde")
plt.show()

x, y = heun_e(5, 0.2)
plt.plot(x,y)
plt.title("Grafiek van de formule e^x Heun")
plt.ylabel("e^x")
plt.xlabel("Gegeven x waarde")
plt.show()
```
