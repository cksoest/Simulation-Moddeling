```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Differentiaal vergelijking</h2>
z`(t) = 0.1 * 6 - (0.1 * 6 + z(t)) / 1000 * 6

```python
def zout_concentratie_tank(steps, h, flow_out):
    t = np.zeros(steps+1)
    z = np.zeros(steps+1)
    t[0] = 0
    z[0] = 0
    for step in range(steps):
        new_z = 0.1*6-(0.1*6+z[step])/1000*flow_out
        z[step+1] = z[step] + h*new_z
        t[step+1] = h * (step+1)
    return t, z
```

<h2>Verloop van de zoutconcentratie in de tank</h2>

```python
t, z = zout_concentratie_tank(8000, 0.1, 6)
plt.plot(t,z)
plt.title("Verloop zout concentratie in de tank")
plt.ylabel("Zout in KG in de tank")
plt.xlabel("Tijd verstreken in minuten")
plt.show()
```

<h2>Verloop van de zoutconcentratie in de tank met dat het water met 5 L/min de tank uitstroomt</h2>

```python
t, z = zout_concentratie_tank(8000, 0.1, 5)
plt.plot(t,z)
plt.title("Verloop zout concentratie in de tank")
plt.ylabel("Zout in KG in de tank")
plt.xlabel("Tijd verstreken in minuten")
plt.show()
```
