```python
import numpy
import matplotlib.pyplot


def forward_euler(spring):
    h=0.1
    g=9.81
    friction = 0.1
    force = -1
    mass = 1
    spring_point = 10
    spring_strenght = -10

    num_steps = 50
    t = numpy.zeros(num_steps+1)
    x = numpy.zeros(num_steps+1)
    v = numpy.zeros(num_steps+1)

    for step in range(num_steps):
        if spring:
            force = spring_strenght * x[step] - spring_point
        acceleration = force/mass
        t[step+1] = h * (step+1)
        v[step+1] = v[step] + acceleration * h
        x[step+1] = x[step] + v[step] * h
    return t,x,v


def plot_me():
    axes_height = matplotlib.pyplot.subplot(211)
    matplotlib.pyplot.plot(t, x)
    axes_velocity = matplotlib.pyplot.subplot(212)
    matplotlib.pyplot.plot(t, v)
    axes_height.set_ylabel('Height in m')
    axes_velocity.set_ylabel('Velocity in m/s')
    axes_velocity.set_xlabel('Time in s')
    matplotlib.pyplot.show()
```

<h2>Grafiek van een vallende bal</h2>

```python
t,x,v = forward_euler(False)
plot_me()
```

<h2>Grafiek van een bal van aan een veer</h2>

```python
t,x,v = forward_euler(True)
plot_me()
```
