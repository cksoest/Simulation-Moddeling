```python
import matplotlib.pyplot as plt
```

<h3>updated random number generator</h3>

```python
import numpy as np

class random_number_generator:

    def __init__(self, seed):
        self.number = seed
        self.a = 1664525
        self.c = 1013904223
        self.m = 2**32
    
    def next_num(self):
        next_num = (self.a*self.number+self.c)%self.m
        self.number = next_num
        return next_num/self.m
    
# toegevoegde code die een random number geeft uit een normaal verdeling met mu en sigma als gegeven constanten.
####################################################################    
    def next_num_norm(self, mu, sigma):
        u1 = self.next_num()
        u2 = self.next_num()
        z1 = np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
        return z1*sigma+mu
####################################################################
```

<h3>Functie die een normaal verdeling vergeeft op basis van de methode "next_num_norm" van de random number generator</h3>

```python
def draw_hist_random_norm(num_steps, bins, mu, std):
    rand_norm = random_number_generator(52)
    data = []
    for i in range(num_steps):
        u1 = rand_norm.next_num()
        u2 = rand_norm.next_num()
        z1 = np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
        x = z1*std+mu
        data.append(x)
    plt.hist(data, bins=bins)
    plt.title("normaalverdeling van de random number generator")
    plt.xlabel("Waarde van X")
    plt.ylabel("Aantal keer dat X voorkomt")
    plt.show()
```

<h3>Normaal verdeling met 100 waardes</h3>

```python
draw_hist_random_norm(100, 200, 0, 1)
```

<h3>Normaal verdeling met 1000 waardes</h3>

```python
draw_hist_random_norm(1000, 200, 0, 1)
```

<h3>Normaal verdeling met 10000 waardes</h3>

```python
draw_hist_random_norm(10000, 200, 0, 1)
```

<h3>Normaal verdeling met 100000 waardes</h3>

```python
draw_hist_random_norm(100000, 200, 0, 1)
```
