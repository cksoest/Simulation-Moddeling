```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Differentiaal vergelijkingen</h2>
z1`(t) = 0.2*6 + z2(t)/100 - z1(t)/100*7<br>
z2`(t) = z1(t)/100*3 - z2(t)/100*3

```python
def zout_concentratie_tank(steps, h):
    t = np.zeros(steps+1)
    z1 = np.zeros(steps+1)
    z2 = np.zeros(steps+1)
    
    t[0] = 0
    z1[0] = 0
    z2[0] = 20
    
    for step in range(steps):
        new_z1 = 0.2*6 + z2[step]/100 - z1[step]/100*7
        new_z2 = z1[step]/100*3 - z2[step]/100*3
        z1[step+1] = z1[step] + h*new_z1
        z2[step+1] = z2[step] + h*new_z2
        t[step+1] = h * (step+1)
    return t, z1, z2
```

```python
t, z1, z2 = zout_concentratie_tank(2000, 0.1)
plt.plot(t,z1)
plt.plot(t,z2)
plt.title("Verloop zout concentratie in de tanks")
plt.ylabel("Zout in KG in de tank")
plt.xlabel("Tijd verstreken in minuten")
plt.legend(["tank1", "tank2"])
plt.show()
```
